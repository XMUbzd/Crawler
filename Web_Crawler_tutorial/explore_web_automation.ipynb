```python
from requests_html import HTMLSession
session = HTMLSession()
r = session.get('https://k.autohome.com.cn/')

# r.html.render()
sel = 'body > div > div.row > div.column.grid-13 > div:nth-child(1) > div.result > ul > li:nth-child(1) > div.car-tags > span:nth-child(3)'

print(r.html.find(sel, first=True).text)

# r.html.xpath('a')
```

```python
import requests

response = requests.get('https://k.autohome.com.cn/')
response.encoding = 'utf-8'
response.text
with open('autohome.html',encoding='utf8',mode="w") as f:
    f.write(response.text)
```

```python
with open('autohome.html',encoding='utf8', mode="r") as f:
    html = f.read()
```
